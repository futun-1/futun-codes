### Check if the gender affects the loan approval

```python
Gender=pd.crosstab(df['Gender'],df['Loan_Status'], normalize='index')
print(Gender)
Gender.plot(kind='bar', stacked=True)
```

```python
# Calculate approval rates for each gender group
gender_approval_rates = df.groupby('Gender')['Loan_Status'].value_counts(normalize=True).unstack()
# Compare approval rates
approval_rate_difference = gender_approval_rates['Y']['Male'] - gender_approval_rates['Y']['Female']

# Print the difference in approval rates
print("\nDifference in Approval Rates (Female - Male):", approval_rate_difference)
```

According to the difference between approved females and males, gender has an effect on approval rates of 0.02

```python
# Calculate proportions for better comparison
normalized_data = Gender.div(Gender.sum(axis=1), axis=0)

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(normalized_data, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Relationship between Gender and Loan Status')
plt.xlabel('Loan_Status')
plt.ylabel('Gender')
plt.show()
```

# 5. Preprocessing
"Garbage in, garbage out".
<br>
<br>
Data should be preprocessed and cleaned to get rid of noisy data. Preprocessing includes:

- remove unsued column(s)
- remove grabage data
- convert continous data into categorical
- treat missing data
- encode categorical data
- normalization: normalize data ranges for all features (e.g., between 0 and 1)

## Remove unused columns
Here we will drop some columns we beleive they do not affect the loan approval.

```python
sns.pairplot(df, x_vars=['Loan_ID','Gender','Self_Employed','ApplicantIncome','LoanAmount','Loan_Amount_Term'], y_vars='Loan_Status', kind='scatter')
plt.show()
df.drop(['Loan_ID','Gender','Self_Employed','ApplicantIncome','LoanAmount','Loan_Amount_Term'], inplace=True, axis=1)
```

##Convert continous data into categorical

```python
df['Credit_History'] = df['Credit_History'].astype('O')
```

## Treat missing data
- fill missing categorical data & discrete data with mode
- fill missing continous data with mean


```python
df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
df['Married'].fillna(df['Married'].mode()[0], inplace=True)
```

## Check if any null value exist

check for any null values

```python
df.isnull().sum()
```

## Encode categorical data

```python
le = LabelEncoder()
cat_data=df[['Credit_History','Dependents','Education','Married','Property_Area','Loan_Status']]
num_data=df[['CoapplicantIncome']]
cat_data.head()
```

```python
for i in cat_data:
    cat_data[i] = le.fit_transform(cat_data[i])
cat_data.head()
```

## Combine both cat_data and numerical_data

```python
df = pd.concat([cat_data, num_data], axis=1)
df.head()
```

## Normalization

Normalize data ranges for all features (e.g., between 0 and 1)

```python
# get the max of each column
df_max = df.max()
df_max
# divide each column by its maximum value
df = df.divide(df_max)
df.describe()
```

# 6. Training & Validating

We split the data between training and testing.
Then, we train a model to predict the loan status

```python
input_variables = ['Credit_History','Dependents','Education','Married','Property_Area','CoapplicantIncome']
x = df[input_variables]
y = df['Loan_Status']

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size =0.3)
```

##Creating & Fitting the model

```python
model=LogisticRegression()
model.fit(x_train, y_train)
```

##Validating the model

```python
pred_val = model.predict(x_val)
accuracy_score(y_val,pred_val)
```
